# Motor Trend Car Road Tests

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_excel("Motor Trend Car Road Tests.xlsx")
data
```

```python
y = data["mpg"]
x = data.drop(columns=["mpg", "model"])
```

```python
n = len(data)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.4, random_state = 137)
```

## Escalamiento

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler().fit(x_train)
X_train_scaled = scaler.transform(x_train)
```

```python
lr = LinearRegression()
lr.fit(X_train_scaled, y_train)
```

### Train

```python
y_hat = lr.predict(X_train_scaled)
r2_score(y_train, y_hat)
```

```python
n = len(X_train_scaled)
unos = np.ones([n,1])
X = np.column_stack([unos,X_train_scaled])
ols = sm.OLS(y_train, X_train_scaled)
results = ols.fit()
results.summary()
```

### Test

```python
X_test_scaled = scaler.transform(x_test)
```

```python
lr = LinearRegression()
lr.fit(X_test_scaled, y_test)
```

```python
y_hat_test = lr.predict(X_test_scaled)
r2_score(y_test, y_hat_test)
```

```python
n = len(X_test_scaled)
unos = np.ones([n,1])
X = np.column_stack([unos,X_test_scaled])
ols = sm.OLS(y_test, X_test_scaled)
results = ols.fit()
results.summary()
```
