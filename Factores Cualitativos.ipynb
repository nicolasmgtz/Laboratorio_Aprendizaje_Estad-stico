# Factores Cualitativos

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_excel("Tabla Factores Cualitativos.xlsx")
data
```

```python
x = data[["Kahlua", "Coco", "Horchata", "Mandarina", "Tamaño", "Hielo" ]]
```

```python
y = data["Calificación"]
```

```python
n = len(y)
```

```python
unos = np.ones([n,1])
```

```python
X = np.column_stack([unos,x])
```

```python
ols = sm.OLS(y, X)
```

```python
results = ols.fit()
```

```python
results.summary()
```

```python
data = pd.read_excel("Tabla Factores Cualitativos.xlsx", sheet_name="Hoja2")
```

```python
x = data[["Kahlua", "Coco", "Horchata", "Mandarina", "Tamaño", "Hielo" ]]
y = data["Calificación"]
n = len(y)
unos = np.ones([n,1])
X = np.column_stack([unos,x])
ols = sm.OLS(y, X)
results = ols.fit()
results.summary()
```
