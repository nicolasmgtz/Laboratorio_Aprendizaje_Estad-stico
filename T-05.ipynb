## 1.1 Realiza una regresión tomando 'mpg' como salida y eliminando la columna 'model'. Considera todos los demás factores como numéricos/ordinales.

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso
```

```python
data = pd.read_excel("Motor Trend Car Road Tests.xlsx")
```

```python
y = data["mpg"]
x = data.drop(columns=["mpg", "model"])
```

```python
n = len(data)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.4, random_state = 137)
```

### Escalamiento

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler().fit(x_train)
X_train_scaled = scaler.transform(x_train)
```

```python
lr = LinearRegression()
lr.fit(X_train_scaled, y_train)
```

### Train

```python
y_hat = lr.predict(X_train_scaled)
r2_score(y_train, y_hat)
```

```python
n = len(X_train_scaled)
unos = np.ones([n,1])
X = np.column_stack([unos,X_train_scaled])
ols = sm.OLS(y_train, X)
results = ols.fit()
results.summary()
```

##### Interpretación signos betas

- const = 17.4333. Si todos los predictores se fijaran en cero, el modelo estima que el auto tendría un rendimiento de aproximadamente 17.43 millas por galón.

- x1 = 40.7571. Cuando aumenta este predictor, el valor de mpg crece en 40.76 unidades, indicando una relación positiva fuerte con el rendimiento de combustible.

- x2 = -2.4635. Un incremento en esta variable hace que el mpg disminuya en 2.46 unidades, lo que refleja mayor consumo.

- x3 = -10.1581. Si este predictor aumenta, el mpg baja en 10.16 unidades, mostrando un efecto negativo importante sobre la eficiencia.

- x4 = 14.1135. Un aumento en este valor se asocia con un incremento de 14.11 unidades en el mpg, mejorando el rendimiento.

- x5 = -0.8005. Cuando crece este factor, el mpg se reduce en 0.80 unidades.

- x6 = 1.5727. Este predictor se relaciona con un incremento de 1.57 unidades en el mpg, reflejando mejor eficiencia.

- x7 = 32.4926. Al aumentar esta variable, el mpg crece en 32.49 unidades, lo que implica un efecto positivo fuerte.

- x8 = 36.1394. Si este valor sube, el mpg se incrementa en 36.14 unidades.

- x9 = -38.3551. Un aumento en esta variable reduce el mpg en 38.36 unidades, indicando un efecto muy negativo sobre la eficiencia.
